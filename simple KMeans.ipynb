```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.random.uniform(-100, 100, 100)
Y = np.random.uniform(-100, 100, 100)
```

```python
Y
```

```python
X
```

```python
arr = []
for i in X:
  for j in Y:
    a = [i ,j]
    arr.append(a)
```

```python
count = 0
for i in arr:
  if count % 100 not in range(0,10):
    arr.remove(i)
  count += 1
```

```python
len(arr)
```

```python
import random

final_datapoints_count = (len(arr) * 0.0275)
random_indexes = random.sample(range(len(arr)), int(final_datapoints_count))
sparse_arr = [arr[i] for i in random_indexes]
```

```python
final_datapoints_count
```

```python
x, y = zip(*sparse_arr)
```

```python
plt.scatter(x,y)
```

```python
from sklearn.cluster import KMeans
```

```python
from sklearn.cluster import KMeans
import numpy as np

sparse_arr_np = np.array(sparse_arr)

n_clusters = 3

kmeans = KMeans(n_clusters=n_clusters, random_state=10, n_init=300)

kmeans.fit(sparse_arr_np)

labels = kmeans.labels_

centers = kmeans.cluster_centers_

print("Cluster labels:", labels)
print("Cluster centers:", centers)
```

```python
# Visualize the clusters
plt.figure(figsize=(8, 6))
plt.scatter(sparse_arr_np[:, 0], sparse_arr_np[:, 1], c=labels, cmap='viridis', marker='o', s=20, alpha=0.6)
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.9, marker='X', label='Cluster Centers')
plt.title('K-Means Clustering Results')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()
```
